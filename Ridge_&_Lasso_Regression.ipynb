```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/50_Startups.csv')
df.head()
```

```python
df.shape
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df.State = le.fit_transform(df.State)
```

```python
df.head()
```

## X and y split


```python

X = df.drop(columns = ['Profit'],axis =1)
X.head()
```

```python
y =df.Profit
y.head()
```

## Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scale =MinMaxScaler()
```

```python
X_scaled= pd.DataFrame(scale.fit_transform(X),columns =X.columns)
X_scaled.head()
```

## Train test split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

## Model building

```python
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
r = Ridge()
l = Lasso()
```

```python
r.fit(x_train,y_train)
```

```python
l.fit(x_train,y_train)
```

```python
pred1 =r.predict(x_test)
```

```python
pred1
```

```python
pred1_train = r.predict(x_train)
```

```python
pred1_train
```

```python
pred2 =l.predict(x_test)
```

```python
pred2
```

```python
pred2_train = l.predict(x_train)
pred2_train
```

```python
profit =pd.DataFrame({'Actual_Profit':y_test, 'Ridge_pred':pred1,'Lasso_pred':pred2})
profit.head(10)
```

## Evaluating Ridge and Lasso models

```python
from sklearn import metrics
```

```python
# r2 score
print(metrics.r2_score(y_test,pred1))
print(metrics.r2_score(y_test,pred2))
```

```python
# MSE
print(metrics.mean_squared_error(y_test,pred1))
print(metrics.mean_squared_error(y_test,pred2))
```

```python
# RMSE

print(np.sqrt(metrics.mean_squared_error(y_test,pred1)))
print(np.sqrt(metrics.mean_squared_error(y_test,pred2)))
```

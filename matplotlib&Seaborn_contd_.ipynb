```python
# pie chart
data =[2,6,5,10]
plt.pie(data)
plt.show()
```

```python
plt.pie(data,[0,0,0.1,0],labels=['t1','t2','t3','t4'],autopct = "%1.1f%%",shadow=True,colors = ['yellow','k','red','green'])
plt.title('Pie diagram')
plt.show()
```
